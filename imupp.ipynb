```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data_imupp.csv")

df
```

```python
df['datetime'] = pd.to_datetime(df['datetime'], dayfirst=True)
df
```

```python
df.info()
```

```python
y = 'humidity'
x_cols = ['tempmax','tempmin','temp','dew','precip',
          'windgust','windspeed','winddir','sealevelpressure',
          'cloudcover']  # tanpa precipprob
```

```python
corr = df[x_cols + [y]].corr()
```

```python
#Korelasi Pearson

corr_y = corr[y].drop(y).sort_values(ascending=False)
print("\nKorelasi tiap variabel terhadap Kelembapan (humidity):")
print(corr_y)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr[x_cols + [y]], annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Variabel Cuaca terhadap Humidity')
plt.show()
```

Berdasarkan heatmap korelasi di atas, variabel yang paling berpengaruh terhadap kelembapan udara adalah cloudcover (r = 0.71) dan precipitation (r = 0.39). Sementara variabel suhu (temp, tempmax, tempmin) menunjukkan korelasi yang sangat rendah (sekitar 0.09–0.12), sehingga pada dataset ini suhu tidak menjadi faktor dominan dalam perubahan kelembapan udara.


```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x=df['tempmax'] + np.random.normal(0, 0.1, len(df)),
    y=df['humidity'] + np.random.normal(0, 0.1, len(df)),
    alpha=0.6
)
sns.regplot(x='tempmax', y='humidity', data=df, scatter=False, color='red', line_kws={'lw':1})
plt.title('Hubungan tempmax terhadap Humidity')
plt.xlabel('tempmax')
plt.ylabel('humidity')
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['datetime'], df['humidity'], marker='o')
plt.title('Trend Curah Hujan Harian')
plt.xlabel('Tanggal')
plt.ylabel('Humidity')
plt.grid(True)
plt.show()
```

Berdasarkan grafik tren curah hujan harian yang dianalisis, terlihat adanya pola penurunan kelembapan yang konsisten dari tahun 2021 sampai 2025. Pada awal tahun 2021 kelembapan mencapai nilai tertinggi sekitar 95, namun secara bertahap menurun dan diproyeksikan mendekati 55 pada tahun 2023.


Tren jangka panjang ini mengindikasikan atmosfer yang semakin kering. Tren penurunan kelembapan ini berpotensi mengurangi curah hujan, meningkatkan resiko kekeringan, serta mempengaruhi sektor pertanian dan ketersediaan air.

```python
print("\n=== Interpretasi Korelasi ===")
for var, val in corr_y.items():
    if val > 0.8:
        status = "Sangat kuat positif"
    elif val > 0.6:
        status = "Kuat positif"
    elif val > 0.4:
        status = "Sedang"
    elif val > 0.2:
        status = "Rendah positif"
    elif val > 0:
        status = "Sangat Rendah Positif"
    elif val > -0.3:
        status = "Lemah Negatif"
    elif val > -0.8:
        status = "Sangat Lemah Negatif"
    else:
        status = "tidak berkorelasi signifikan"
    print(f"{var:20s}: {val:.3f} → {status}")
```
